```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

import pickle

model = pickle.load(open('modelo_mycnn_10_71.pkl', 'rb'))

# Load and preprocess the test image
image_path = 'archive/Fruit-262/grapefruit/255.jpg'  # Replace with the actual path to the test image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (32, 32))
image = np.expand_dims(image, axis=0)
image = image / 255.0

# Make predictions
predictions = model.predict(image)
predicted_class_index = np.argmax(predictions[0])

# Define class labels (modify based on your model's classes)
class_labels = ['acai', 'apple', 'avocado', 'banana', 'blueberry', 'carambola', 'coconut', 'coffee', 'durian', 'grapefruit', 'jaboticaba', 'kiwi']

# Display the image and predicted class
plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Predicted class: ' + class_labels[predicted_class_index])
plt.show()
```
